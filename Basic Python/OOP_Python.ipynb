Object-Oriented Programming (OOP) is a programming paradigm that organizes code into objects, which are instances of classes. It allows for better code organization, reusability, and easier maintenance.

In Python, OOP is implemented using classes and objects. Here's a breakdown of the key concepts in OOP and how they are implemented in Python:

# 1. Classes and Objects
Class: A blueprint or template for creating objects.


````python
class Car:
    # Constructor (initializer) to create an object of the class
    def __init__(self, make, model, year):
        self.make = make  # Attribute
        self.model = model  # Attribute
        self.year = year  # Attribute

    # Method (function within a class)
    def start_engine(self):
        print(f"The engine of the {self.year} {self.make} {self.model} is now running.")

# Creating an object (instance) of the Car class
my_car = Car("Toyota", "Corolla", 2020)
my_car.start_engine()  # Calling the method
````

# **3**. Inheritance
Inheritance allows one class to inherit the attributes and methods from another class. This helps in code reuse and creating hierarchical relationships between classes.

Example:

````python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement abstract method")

# Inheriting Animal class
class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"

# Creating objects of Dog and Cat
dog = Dog("Buddy")
cat = Cat("Whiskers")

print(dog.speak())  # Buddy says Woof!
print(cat.speak())  # Whiskers says Meow!
````

# *4*. Polymorphism
Polymorphism allows different classes to have methods with the same name, but each class can have its own implementation of the method.

Poly = Many

Example:


````python
class Bird(Animal):
    def speak(self):
        return f"{self.name} says Chirp!"

# Creating objects of Bird
bird = Bird("Tweety")

# Demonstrating polymorphism
animals = [dog, cat, bird]
for animal in animals:
    print(animal.speak())
````

# 2. Encapsulation
Encapsulation is the concept of bundling the data (attributes) and methods (functions) that operate on the data within a single unit (class), and restricting access to some of the object's components (often by using private attributes).

In Python, we can achieve encapsulation by prefixing attributes or methods with an underscore (_) for protected access, or double underscore (__) for private access.

Access Modifier:<br>
**Public <br>
Privet <br>
Protected <br>**

````python
#Public
class A:
  def __init__(self,name):
    self.name = name
obj1=A("Nazmul")
print(obj1.name)
````

````python
#Protcted (Data Secure) Can't use without object
class A:
  def __init__(self,name): #protcted
    self._name = name
obj1=A("Nazmul")
print(obj1._name)
````

````python
#Privet
class A:
  def __init__(self,name):
    self.__name = name #privet
obj1=A("Nazmul")
print(obj1.__name)
````

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age  # Private attribute

    # Getter method for private attribute
    def get_age(self):
        return self.__age

    # Setter method for private attribute
    def set_age(self, age):
        if age > 0:
            self.__age = age

# Creating an object of Person class
person = Person("Alice", 30)

# Accessing private attribute via getter
print(person.get_age())  # 30

# Modifying private attribute via setter
person.set_age(35)
print(person.get_age())  # 35
````

# 5. Abstraction
Abstraction is the concept of hiding the complex implementation details and showing only the essential features of the object. In Python, we can use abstract classes and methods (via the abc module) to achieve abstraction.

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * (self.radius ** 2)

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Creating objects of Circle and Rectangle
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Circle Area: {circle.area()}")
print(f"Rectangle Area: {rectangle.area()}")
````
