## Example: Using file.write(), file.seek(), and file.read()

```python
# Open the file in write mode to add some content
with open('file.txt', 'w') as file:
    # Writing a string to the file
    file.write("Hello, this is the first line.\n")
    file.write("This is the second line.\n")

# Open the file again in read-write mode
with open('file.txt', 'r+') as file:
    # Move the pointer to the beginning of the file
    file.seek(0)

    # Read the content after seeking to the beginning
    print("Reading from the start of the file:")
    print(file.read())  # This will print the entire content from the start

    # Now, move the pointer to a specific position (e.g., after the first line)
    file.seek(10)  # Moves the pointer to the 10th byte

    # Read 10 bytes from the current position (after seeking)
    print("\nReading after seeking to the 10th byte:")
    print(file.read(10))  # Reads 10 characters starting from byte 10
```

Read line by line

```python
with open('file.txt', 'r') as file:
    for line in file:
        print(line.strip())  # .strip() removes the newline character at the end of each line
```

# Using r+ mode (Read & Write)

```python
# Using r+ mode (Read & Write)
with open("example.txt", "r+") as file:
    content = file.read()
    print("Before writing:", content)
    file.write("\nUpdated content!")
    print(content)
```
