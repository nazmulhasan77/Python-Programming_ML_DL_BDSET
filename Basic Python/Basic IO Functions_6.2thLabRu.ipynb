## 1. Opening and Closing Files
Python provides the `open()` function to work with files.
**Syntax:**
`file = open("filename.txt", "mode")`

**Modes:**
- "r": Read mode (default) – file must exist
- "w": Write mode – creates a new file or overwrites existing content
- "a": Append mode – adds new data at the end of the file
- "r+": Read and Write – file must exist
- "w+": Write and Read – creates/overwrites file
- "a+": Append and Read – creates file if it doesn’t exist


## 2. Closing a File
`file.close()` ensures changes are saved and resources are freed.

### Why is Closing a File Important?
- Prevents data loss & corruption
- Python buffers file operations, changes might be stored in memory before being written to disk

### ⚡ Best Practice: Using `with` Statement
```python
with open("data.txt", "r") as file:
    content = file.read()
    print(content)
# No need to close the file manually
```

````python
# Exercise 1: Creating & Writing to a File
with open("example.txt", "w") as file:
    file.write("Hello, this is a test file.\n")
    file.write("We are learning Python file handling.")
````

````python
# Exercise 2: Reading from a File (First 10 characters)
with open("example.txt", "r") as file:
    content = file.read(10)
    print("First 10 characters:", content)
````

````python
# Exercise 3: Appending Data to a File
with open("example.txt", "a") as file:
    file.write("\nAppending new information.")

# Verify the content
with open("example.txt", "r") as file:
    print(file.read())
````

````python
# Exercise 4: Reading a File Line by Line, print lines containing 'Python'
with open("example.txt", "r") as file:
    for line in file:
        if "Python" in line:
            print(line.strip())
````

````python
# Exercise 5: Using r+ mode
with open("example.txt", "r+") as file:
    content = file.read()
    print("Before writing:", content)
    file.write("\nUpdated content!")
````

## Summary Table: Comparing File Modes
| Mode | Read | Write | Overwrites File? | Appends to End? | Cursor Position | Creates File? |
|------|------|-------|------------------|------------------|------------------|----------------|
| r+   | ✅   | ✅    | ❌               | ❌               | Start            | ❌             |
| w+   | ✅   | ✅    | ✅               | ❌               | Start            | ✅             |
| a+   | ✅   | ✅    | ❌               | ✅               | End              | ✅             |

````python
# Exercise 6: Using os module
import os
file_path = "/content/sample.txt"
if os.path.exists(file_path):
    print("File exists. Deleting now...")
    os.remove(file_path)
else:
    print("File does not exist.")
````

````python
# Exercise 6: Using pathlib module
from pathlib import Path
file_path = Path("/content/sample.txt")
if file_path.exists():
    print("File exists. Deleting now...")
    file_path.unlink()
else:
    print("File does not exist.")
````
