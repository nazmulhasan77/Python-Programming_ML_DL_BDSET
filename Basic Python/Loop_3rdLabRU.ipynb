Reverse string

```python
string=input("Enter a string: ")
reverseString=""
for char in string:
  reverseString=char+reverseString
print("Reverse string is: ",reverseString)
```

2. Print average of all numbers divisible by 3 and less than 100.

```python
sum=0
count=0
for i in range(1,100):
  if(i%3==0):
    sum=sum+i
    count=count+1
print("Average of all numbers divisible by 3 and less than 100: ",sum/count)
```
