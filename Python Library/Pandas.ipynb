#Selecting Columns and Rows
[link text](https://)

```python
df['Name']  # Select the 'Name' column
df[['Name', 'Age']]  # Select 'Name' and 'Age' columns
df.iloc[0]  # Select the first row
df.iloc[1:3]  # Select rows 1 and 2 (Python slicing)
df[df['Age'] > 30]  # Select rows where Age > 30
```

#Adding and Modifying Columns

```python
df['Country'] = ['USA', 'France', 'Germany', 'UK'] #add new column
df
df['Age'] = df['Age'] + 1  # Increment all ages by 1
df
```

#Dropping Columns and Rows

```python
df.drop('Country', axis=1, inplace=True)  # Drop column 'Country'
df.drop(0, axis=0, inplace=True)  # Drop the first row (index 0)
```

# Sorting Data

```python
df.sort_values(by='Age', ascending=False)  # Sort by Age in descending order
df.sort_values(by=['Age', 'Name'], ascending=[True, False])  # Sort by Age (asc), then Name (desc)
```

#Handling Missing Data

```python
df.isnull()  # Returns True if value is missing, else False
df.dropna(inplace=True)  # Drop rows with any NaN values
df['Age'].fillna(df['Age'].mean(), inplace=True)  # Replace NaN in 'Age' with the mean value
```
