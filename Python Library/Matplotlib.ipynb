# Scatter Plot

```python
x = [1,2,3,4,5,6,7]
y = [8,9,10,11,12,13,14]
```

```python
plt.scatter(x,y,color='red',marker='h')
plt.plot(x,y,ls='--',color='green') # connect line
```

```python
plt.figure(figsize=(10,5) ) # canvas size
plt.scatter(x,y,color='red',marker='h',label='Scatter plot')
plt.plot(x,y,ls='--',color='green',label='line plot') # connect line
plt.legend(loc='best',shadow=True) # show level , set location
plt.grid(ls='--',alpha=0.5) # grid style , opacity
plt.xlabel('X-axis',color='red')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.savefig('scatter.png') # save image
plt.show() # render view
```

# **Wave**

## Sine Wave

```python
x=np.linspace(0,4*np.pi,1000)
y=np.sin(x)
plt.plot(x,y,color='green')
plt.axhline(y=0,color='red') #horizontal line
plt.axvline(x=0,color='red') #vertical line
```

## Cos wave , Cos+Sin wave

```python
x=np.linspace(0,4*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
y3=y1+y2
plt.plot(x,y1,color='green',label='sin wave',ls='--')
plt.plot(x,y2,color='red',label='cos wave',ls='--')
plt.plot(x,y3,color='blue')
plt.legend(loc='best')
plt.axhline(y=0,color='red') #horizontal line
plt.axvline(x=0,color='red') #vertical line
```

## Parabola

```python
import numpy as np
import matplotlib.pyplot as plt

#x = np.linspace(-10, 10, 4)  # np.linspace(start, stop, num) , Generates 400 points between -10 and 10
x = np.arange(-10,10,.2) #np.arange(start, stop, step)

# Define the quadratic function (parabola)
a, b, c = 1, 0, 0  # Coefficients for y = ax^2 + bx + c
y = a * x**2 + b * x + c

# Plot the parabola
plt.figure(figsize=(8, 6))
plt.plot(x, y,label='Parabola')

# Formatting the plot
plt.axhline(0, color="black", linewidth=1)  # X-axis
plt.axvline(0, color="black", linewidth=1)  # Y-axis
plt.grid(True, linestyle="--", alpha=0.6)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Parabola Plot")
plt.legend()

# Show the plot
plt.show()
```

# 3d Plot

```python
x = [1,2,3,4,5,6]
y = [7,8,9,10,11,12]
z =[9,10,11,12,13,14]
```

```python
fig = plt.figure(figsize=(4,5)) # Create a figure with specified size (4 inches wide and 5 inches tall)

ax = fig.add_subplot(111, projection='3d') # Add a 3D subplot to the figure (1 row, 1 column, 1 plot)

ax.scatter(x, y, z)  # Plot a 3D scatter plot using the x, y, and z data points
# Set label for the X-axis
ax.set_xlabel('X-axis')
# Set label for the Y-axis
ax.set_ylabel('Y-axis')
# Set label for the Z-axis
#ax.set_zlabel('Z-axis')
# Set the title for the plot
ax.set_title('3D Scatter Plot')

plt.show() # Display the plot
```

# Sub Plot

## Using plt.show()

```python
x=np.linspace(0,4*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
y3=y1+y2
plt.plot(x,y1, label='Sin wave',color='r')
plt.axhline()
plt.axvline()
plt.legend()
plt.title('Sin wave')
plt.show() #render plot

plt.plot(x,y2,'g--',label='cos wave')
plt.legend(loc='best')
plt.axhline()
plt.axvline()
plt.title('Cos wave')
plt.show()
```

## Using Axis

```python
x=np.linspace(0,4*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
fig=plt.figure(figsize=(10,5))

ax1=fig.add_subplot(2,2,1)
ax1.plot(x,y1, label='Sin wave',color='r')
plt.axhline()
plt.axvline()
plt.legend()
plt.title('Sin wave')
ax2=fig.add_subplot(2,2,4)
ax2.plot(x,y2,'g--',label='cos wave')
plt.legend(loc='best')
plt.axhline()
plt.axvline()
plt.title('Cos wave')
plt.show()
```

# Bar Plot
Categorical Data Comparison

```python
course=['CSE','EEE','ICE','Math']
fees= [5000,20000,30000,1422]
fig=plt.figure(figsize=(7,7))
ax1=fig.add_subplot(1,1,1)
bars=ax1.bar(course,fees,color=['red','green','orange','black'],label=['CSE','EEE','ICE','MATH'])
ax1.bar_label(bars)
plt.title('Course fees')
plt.xlabel('Course')
plt.ylabel('Fees')
plt.legend()
plt.show()
```
